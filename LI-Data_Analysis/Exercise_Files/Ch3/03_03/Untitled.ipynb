```python
words = open('words', 'r').readlines()
```

```python
words
```

```python
wordclean = [word.strip().lower() for word in sorted(words)]
```

```python
wordclean
```

```python
import collections
```

```python
lengthlist = collections.defaultdict(list)
for word in wordclean:
    lengthlist[len(word)].append(word)
```

```python
lengthlist.keys()
```

```python
def sorted_word(myword):
    return ''.join(sorted(myword))        
```

```python
def anagram(myList):
    anagram_list = collections.defaultdict(list)
    for word in myList:
        for another_word in myList:
            if sorted_word(word) == sorted_word(another_word):
                anagram_list[word].append(another_word)
    return anagram_list
        
```

```python
answerDict = {}
for k,v in lengthlist.items():
    lengthlist[k] = anagram(v)
```
