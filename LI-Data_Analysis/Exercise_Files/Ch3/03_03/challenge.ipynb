```python
import collections
```

```python
lengthlist = collections.defaultdict(list)
for word in wordclean:
    lengthlist[len(word)].append(word)
```

```python
def sorted_word(myword):
    return ''.join(sorted(myword))        
```

```python
def anagram(myList):
    anagram_list = collections.defaultdict(list)
    for word in myList:
        for another_word in myList:
            if sorted_word(word) == sorted_word(another_word):
                anagram_list[word].append(another_word)
    return anagram_list
        
```

```python
def answer(aDict):
    answer = collections.defaultdict(list) 
    for k,v in aDict.items():
        answer[k].append(len(anagram(v)))
    return answer
```

```python
def anagram_2(myword):
    return [word for word in wordclean if sorted_word(word) == sorted_word(myword)]
```
