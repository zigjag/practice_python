```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
```

```python
x
```

```python
sinx = np.sin(x)
```

```python
sinx
```

```python
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
pp.plot(x, sinx)
pp.plot(x, cosx)
```

```python
pp.plot(x, sinx, 'x')
pp.plot(x, cosx, 'o')
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2
pp.plot(x, y)
pp.plot(x,z)

pp.legend(['sin(x) cos(x)', 'cos(x)^2 - sin(x)^2'])
```

```python
np.dot(sinx,cosx)
```

```python
mat_outer = np.outer(sinx,cosx)
mat_outer
```
